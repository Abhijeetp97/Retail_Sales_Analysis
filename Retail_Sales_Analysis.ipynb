Retail sales

```python
import pandas as pd
df_retail = pd.read_csv('/content/retail_sales_dataset.csv')

df_retail
```

```python
print(df_retail.info())
print(df_retail.describe())
```

```python
print(df_retail.isnull().sum())
```

```python
# Sales by product category
sales_by_category = df_retail.groupby('Product Category')['Total Amount'].sum()

print(sales_by_category)
```

```python
# Customer demographics
gender_distribution = df_retail['Gender'].value_counts()

print(gender_distribution)
```

```python
age_distribution = df_retail['Age'].describe()

print(age_distribution)
```

```python
# Top customers
top_customers = df_retail.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)

print(top_customers)
```

```python
# Sales trends over time
df_retail['Date'] = pd.to_datetime(df_retail['Date'])

sales_trend = df_retail.groupby(df_retail['Date'].dt.to_period('M'))['Total Amount'].sum()

print(sales_trend)
```

```python
# Average order value
average_order_value = df_retail['Total Amount'].mean()

print(f"The average order value is: {average_order_value}")
```
